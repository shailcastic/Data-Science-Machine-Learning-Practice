```python
df1 = pd.DataFrame(np.array([['a',5,10],['b',4,61],['c',24,19]]),columns=['name','attr1','attr2'])
```

df1


```python
df1
```

```python
df2 = pd.DataFrame(np.array([['a',12,19],['b',4,65],['c',15,17]]),columns=['name','attr1','attr2'])
```

df2

```python
df2
```

```python
df3 = pd.DataFrame(np.array([['a',15,10],['b',14,61],['e',54,69]]),columns=['name','attr1','attr2'])
```

```python
df3
```

```python
pd.merge(df1,df2,on='name')   # innerjoin
```

```python
pd.merge(pd.merge(df1,df2,on='name'),df3, on='name') #three table joins
```

```python
df1.merge(df2,on='name').merge(df3, on='name')
```

```python
# how to join multiple columns
pd.merge(df1,df2, how='inner', on=['name','attr1'])
```
