```python
import matplotlib.pyplot as plt  #  this function can be used when it is an array,list,tuple,pandas series
import seaborn as sns  # dataframe
```

```python
sales =[3452,1254,4523,2545]
year =[2015,2016,2017,2018]
profit = [343,452,251,255]
```

```python
plt.plot(year, sales)
```

```python
plt.plot(year, sales,'red') # when we have to declare in the single graph
plt.plot(year, profit,'blue')
plt.title('sales and profit vs years')   # giving title to the chart
plt.xlabel('years')   # labeling the chart
plt.ylabel('sales and profit')
plt.show()
```

```python
plt.subplot(2,1,1)   # when we have to create two charts
plt.plot(year,sales,'red')   
plt.xlabel('years')
plt.ylabel('sales')
plt.title('sales by years')
plt.subplot(2,1,2)
plt.plot(year,profit,'blue')
plt.xlabel('ears')
plt.ylabel('profit')
plt.title("profit by years")
plt.tight_layout()
```

```python
plt.plot(year,sales,'red')
```

```python
plt.plot(year,sales,'red')
plt.xlim((2015,2016))   # zooming the charts for the particular years
plt.ylim((3000,3800))
```

```python
plt.plot(year,sales,'red')
plt.axis((2015,2016,3000,3800))
```

```python
plt.scatter(y=profit,x=sales)  # to make the scatter plot
plt.annotate('customer1',xy=(3333,333))  # to annotate the particular cell 
```
