```python
(flight_data>>select(X.origin,X.dest, X.hour)) #use of piping
```

```python
(flight_data>>drop(X.year, X.month, X.day))
```

```python
(flight_data>>select(~X.hour, ~X.minute))  #~tilda
```

```python
(flight_data>>mask(X.month==1, X.day==1, X.origin=='JFK',X.hour>10))  #mask is used for filtering
```

```python
(flight_data>>arrange(X.distance, X.hour))
```

```python
(flight_data>>arrange(X.distance, X.hour,ascending=False))
```

```python
(flight_data>>mutate(new_distance=X.distance/1000,carrier_origin= X.carrier + X.origin))
```

```python
(flight_data>>group_by(X.origin))
```

```python
(flight_data>>group_by(X.origin)>>summarise(mean_distance=X.distance.mean()))
```

```python
(flight_data>>group_by(X.origin)>>summarise(mean_distance=X.distance.max()))
```

```python
flight_data>>group_by(X.origin)>>summarise(mean_distance=X.distance.min())
```

```python
(flight_data>>mask(X.hour>10)>>mutate(speed=X.distance/(X.air_time*60))>>group_by(X.origin)>>summarise(mean_speed=X.speed.mean())>>
 arrange(X.mean_speed, ascending=False))
```
